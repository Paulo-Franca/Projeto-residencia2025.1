```
%sql
SELECT
    s.nome_aluno,
    s.aluno_id AS id_aluno,
    s.curso_atual AS curso,
    CAST(p.valor_pagamento AS FLOAT) + 
    CAST(p.valor_multa AS FLOAT) - 
    CAST(p.valor_desconto AS FLOAT) AS total_pago,
    p.data_pagamento,
    p.data_vencimento
FROM
    payments_3_csv p
INNER JOIN
    students_updated_4_csv s
    ON CAST(p.aluno_id AS STRING) = CAST(SUBSTRING(s.aluno_id, 3) AS STRING);
```
